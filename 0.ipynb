```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.dpi'] = 100


def edf(samples, x: float) -> float:
    """Эмпирическая функция распределения (empirical distribution
    function).
    """
    return samples[samples < x].shape[0] / samples.shape[0]


def cdf_uniform(a: float, b: float, x: float) -> float:
    """Функция распределения (cumulative distribution function)
    для равномерного распределения.
    """
    if x < a:
        return 0
    elif x < b:
        return (x - a) / (b - a)
    else:
        return 1
    

a = 12
b = 14

samples = np.random.uniform(a, b, size=30)
    
ox = np.linspace(11, 15, 30)
oy1 = np.array([edf(samples=samples, x=x) for x in ox])
oy2 = np.array([cdf_uniform(a=a, b=b, x=x) for x in ox])

plt.plot(ox, oy1, label='edf')
plt.plot(ox, oy2, linestyle='dashed', label='cdf')

plt.legend()
plt.show()

def get_mean_samples_uni(n: int) -> np.ndarray:
    """Возвращает выборку из значений среднего от `n` равномерно распределённых
    случайных величин.
    """
    samples = np.random.uniform(12, 14, size=(n, 30))
#     print(samples.sum(axis=1))
#     return samples.mean(axis=0)
    return samples.sum(axis=1)

n = 2000

samples = get_mean_samples_uni(n)
plt.hist(samples, density=True, bins=30)
plt.show()
samples

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.dpi'] = 100


def get_mean_samples_geom(n: int) -> np.ndarray:
    """Возвращает выборку из значений среднего от `n` геометрически распределённых
    случайных величин.
    """
    samples = np.random.geometric(0.2, size=(n, 10)) - 1
    print(samples)
    return samples.mean(axis=0)

n = 2

samples = get_mean_samples_geom(n)
plt.hist(samples, density=True, bins=30)
plt.show()
samples
```
